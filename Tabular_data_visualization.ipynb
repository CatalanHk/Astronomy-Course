```python
def calc(n1,n2,operation):
    """
    Functions that: add, substract, multiply, divide two input numbers
    ---- Parameters----
    n1 = first number
    n2 = second number
    operation = operation required
    """
    if operation == 'add':
        return(float(n1 + n2))
    elif operation == 'substract':
        return(float(n1 - n2))
    elif operation == 'multiply':
        return(float(n1 * n2))
    elif operation == 'divide':
        return(float(n1 / n2))
    else: 
        print(f'add, substrac, multiply, divide available {operation} non available')
```

```python
ad = calc(9,2,'add')
sb = calc(9,2,'substract')
mt = calc(9,2,'multiply')
dv = calc(9,2,'divide')
wa = calc(9,2,'mesi')
print(ad,sb,mt,dv,wa)
```

## H-R Diagram Creation
- We will create one with our dataset and match it with a real one


Through the data set we will try to create something like this 


```python
# We have Temperature and absolute magnitude of our stars otherwise we cant create hrd
# we fetch the data needed from the dataframe(Temperature and abs mag)
temperature = star_df['Temperature (K)'].values # all the data goes to temperature in an array
abs_mag = star_df['Absolute magnitude(Mv)'].values

# the values in x and y should go from higher to lower
# as magnitudes are brighter when they are negative 
plt.gca().invert_xaxis() #gca means get current axis
plt.gca().invert_yaxis() # we invert both axis


plt.scatter(temperature, abs_mag) # we create a point scatter plot
plt.show()
```

```python
# We want points with different colors depending on the type of star
# different star types with their values associated
star_type = star_df['Star type'].values
print(star_type)
```

```python
# dictionary within the dictionary to include color 
# star label  with their respective color and marker
# hypergiants bigger in value than brown dwarf 

# size is arbitrary.

star_types = {
    0: {'label': 'Brown Dwarf', 'color': 'brown', 'size': 30, 'marker': '.'},
    1: {'label': 'Red Dwarf', 'color': 'red', 'size': 35, 'marker': '.'},
    2: {'label': 'White Dwarf', 'color': 'black', 'size': 40, 'marker': '.'},
    3: {'label': 'Main Sequence', 'color': 'cyan', 'size': 30, 'marker': 'o'},
    4: {'label': 'Supergiants', 'color': 'orange', 'size': 100, 'marker': 'o'},
    5: {'label': 'Hypergiants', 'color': 'maroon', 'size': 150, 'marker': 'o'}
}
```

```python
star_types[0] # here we can see the values associated to brown dwarf
# if we type 'label' or other of the properties we obtain it value
```

```python
# Obtaining values for star number 10. can be applied to every star in dataframe
star_types[star_type[10]]['label'] # we can obtain color, size, marker,label
```

```python
len(star_type) # objects quantity in the star type array
```

```python
# Plotting HR Diagram for T vs M_abs
for i in range(len(star_type)): # for each element in the star type array
    properties = star_types[star_type[i]] # for every star we assign it properties
    plt.scatter(temperature[i], abs_mag[i], # x,y coordinate for every star with T and M
                s = properties['size'], # property assigment in every iteration
                color = properties['color'],
                marker = properties['marker'],
                label = properties['label'])

# axis invertion
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

# Adding data for sun
# Temperature, M_abs, size, color, marker and label
ax_sun = plt.scatter(5778, 4.83, s=75, c='yellow', marker='o', label='Sun')

# get handles (plotted points) and labels from the plot.
# this give 2 lists, the handle points and the labels for them
handles, labels = plt.gca().get_legend_handles_labels()

# This creates a dictionary where the labels and values are the corresponding points. 
# zip creates a list that assign label to every point and takes it to a dict
# dictionary keys must be unique, so repetitive labels gets erased.
# if we dont do this we obtain 240 labels in the legend
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())



plt.xlabel('Temperature [K]')
plt.ylabel('Absolute Magnitude [Mv]')
plt.title('Hertzsprung Rusell Diagram for 240 stars')

plt.savefig(base_dir1+'HR_Diagram')
plt.show()
```
